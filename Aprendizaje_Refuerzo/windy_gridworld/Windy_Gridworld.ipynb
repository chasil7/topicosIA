# Windy Gridworld <br>
<font size= +1>
<p style="text-align: justify">
El problema _Windy Gridworld_ consiste en que se tiene una cadrícula por la cual te puedes desplazar en cuatro direcciones, una celda a la vez. Las acciones que puedes tomar son __"up"__, __"down"__, __"left"__, __"right"__. 
El objetivo es llegar a una celda en especial que será la meta y como es de pensarse, lo más rapido posible, partiendo de una celda. Lo interesante de este problema es que en una zona de la cuadrícula hay viento y con una cantidad de fuerza (que depende de la columna) te impulsará a otra dirección cuando pases por ahí.
</p>

<p style="text-align: justify">A continuación se presentará el ejemplo clásico del GridWorld y la solución a dos problemas que son una variante de éste.Primero definiremos la estructura principal del problema.</p>
</font>

<font size= +1 >
Para el ejemplo y los ejercicios siguientes la estructura de un estado para este problema será el siguiente.
</font>

```julia
struct GWstate <: Estado
    x ::Int64 # posición en el eje horizontal
    y ::Int64 # posición en el eje vertical
end
```

```julia
struct GWaction <:Accion
    a
    function GWaction(a)
        if a!=:up && a!=:down && a!=:left && a!=:right
            error("acción inválida")
        else new(a)
        end
    end
end
```

<font size= +1 >
Cada problema del Gridworld se le asociarán las dimensiones de la cuadrícula, la posición de la celda objetivo, la posicion de la celda de inicio y un arreglo con los pesos del viento.
</font>

```julia
struct GW <:MDP
    rows ::Int64
    columns ::Int64
    goal::GWstate #Tuple{Int64,Int64}
    start::GWstate #Tuple{Int64,Int64}
    wind::Array{Int64,1}
    actions::Array{Accion,1}
    states::Array{GWstate,1}
end
```

<font size= +1 >Funciones básicas del MDP correspondiente al problema</font>

```julia
#terminal
function terminal(modelo::GW,s::GWstate)
    s.x==modelo.goal.x && s.y==modelo.goal.y
end

#Reward
function r( modelo::GW,s::GWstate,a ::Accion, s2::GWstate)
    (terminal(modelo,s)) ? 1 : -1
end

function r( modelo::GW,s::GWstate)
    (terminal(modelo,s)) ? 1 : -1
end

#Regresa una accion aleatoria
function r_accion(mdp::GW)
    rand(mdp.actions)
end

#acciones legales
function aLegales(mdp::GW,s::GWstate)
    
    
    acciones=[]
    #push!(acciones,r_accion(mdp))
    #pop!(acciones)
    if s.y > 1
        push!(acciones,GWaction(:left))
    end
    if s.y<mdp.columns
        push!(acciones,GWaction(:right))
    end
    if s.x >1
        push!(acciones,GWaction(:up))
    end
    if s.x<mdp.rows
        push!(acciones,GWaction(:down))
    end

    acciones
end
```

```julia
function step(gw::GW,s::GWstate, a ::GWaction)
        
    if a.a == :up 
        return GWstate(max(s.x - 1 - gw.wind[s.y], 1), s.y)
    elseif a.a == :down
        return GWstate(max(min(s.x + 1 - gw.wind[s.y], gw.rows), 1), s.y)
    elseif a.a == :left
        return GWstate(max(s.x - gw.wind[s.y], 1), max(s.y - 1, 1))
    elseif a.a == :right
        return GWstate(max(s.x - gw.wind[s.y], 1), min(s.y + 1, gw.columns))
    end
end
```

```julia
function figure_6_3(mdp::GW, episode_limit,ϵ,α,runs)
    
    q_value=Dict((mdp.start,rand(mdp.actions))=>0.0 )
    
    for s ∈ mdp.states
        for a ∈ aLegales(mdp,s)
        q_value[(s,a)]=0.0
        end
    end
    #q_value = Dict( (s,a)=> 0 for s in mdp.states, a in aLegales(s))
    
    π_ ,steps = sarsa(mdp, q_value, mdp.start, ϵ, α, episode_limit, false,runs)
    
    
    
    optimal_policy = []
    
    for i in 1:mdp.rows
        push!(optimal_policy,[])
        for j in 1:mdp.columns
            s=GWstate(i,j)
            if s==mdp.goal 
                push!(optimal_policy[i],'G')
                continue 
            end
            if π_[s].a==:up
                push!(optimal_policy[i],'U')
            elseif π_[s].a==:down
                push!(optimal_policy[i],'D')
            elseif π_[s].a==:right
                push!(optimal_policy[i],'R')
            elseif π_[s].a==:left
                push!(optimal_policy[i],'L')
            elseif π_[s].a==:upR
                push!(optimal_policy[i],"UR")
            elseif π_[s].a==:downR
                push!(optimal_policy[i],"DR")
            elseif π_[s].a==:upL
                push!(optimal_policy[i],"UL")
            elseif π_[s].a==:downL
                push!(optimal_policy[i],"DL")
            end
        end
    end
    
    
    println("Optimal policy is:")
    for row in optimal_policy
        println(row)
    end
    print("[")
    for i in 1:length(mdp.wind)
        w=mdp.wind[i]
        print("$w , ")
    end
    print("]")

    
    
    steps_ = accumulate(+, (steps))
    y = [i for i in 1:length(steps)]
    
    font = Plots.font("Helvetica", 18)
    pyplot(guidefont=font, xtickfont=font, ytickfont=font, legendfont=font)
    plot(steps_,y , label="line")  
    xlabel!("Time steps")
    ylabel!("Episodes")
    title!("steps vs episodios")
    

end
```

```julia
#estados
col=10
ren=7
est=[]
obj=GWstate(4,8)
ini=GWstate(4,1)
v=[0, 0, 0, 1, 1, 1, 2, 2, 1, 0]
acc=[GWaction(:up),GWaction(:down),GWaction(:right),GWaction(:left)]
for i in 1:ren,j in 1:col
    push!(est,GWstate(i,j))
end

mdp=GW(ren,col,obj,ini,v,acc,est)

```

```julia
episode_limit = 500
ϵ= 0.1
α=0.5
runs=1

figure_6_3(mdp, episode_limit,ϵ,α,runs)
```

## Ejercicio 6.9

### Windy Gridworld con movimientos de Rey

Resuelve con 8 posibles movimientos, incluyendo las diagonales.

```julia
struct GWaction8 <:Accion
    a
    function GWaction8(a)
        if a!=:up && a!=:down && a!=:left && a!=:right && a!=:upR && a!=:upL && a!=:downR && a!=:downL
            error("acción inválida")
        else new(a)
        end
    end
end
```

```julia
function step(gw::GW,s::GWstate, a ::GWaction8)
        
    if a.a == :up 
        return GWstate(max(s.x - 1 - gw.wind[s.y], 1), s.y)
    elseif a.a == :down
        return GWstate(max(min(s.x + 1 - gw.wind[s.y], gw.rows), 1), s.y)
    elseif a.a == :left
        return GWstate(max(s.x - gw.wind[s.y], 1), max(s.y - 1, 1))
    elseif a.a == :right
        return GWstate(max(s.x - gw.wind[s.y], 1), min(s.y + 1, gw.columns))
        
    elseif a.a == :upR
        return GWstate(max(s.x  - 1- gw.wind[s.y], 1), min(s.y + 1, gw.columns))
    elseif a.a == :upL
        return GWstate(max(s.x  - 1- gw.wind[s.y], 1), max(s.y - 1, 1))
    elseif a.a == :downR
        return GWstate(max(min(s.x + 1 - gw.wind[s.y], gw.rows), 1), min(s.y + 1, gw.columns))
    elseif a.a == :downL
        return GWstate(max(min(s.x + 1 - gw.wind[s.y], gw.rows), 1), max(s.y - 1, 1))
    end
end
```

```julia
#acciones legales
function aLegales(mdp::GW,s::GWstate)

    acciones=[]

    if s.y > 1
        push!(acciones,GWaction8(:left))
        if s.x>1 push!(acciones,GWaction8(:upL)) end
        if s.x<mdp.rows push!(acciones,GWaction8(:downL)) end
    end
    if s.y<mdp.columns
        push!(acciones,GWaction8(:right))
        if s.x>1 push!(acciones,GWaction8(:upR)) end
        if s.x<mdp.rows push!(acciones,GWaction8(:downR)) end
    end
    if s.x >1
        push!(acciones,GWaction8(:up))
    end
    if s.x<mdp.rows
        push!(acciones,GWaction8(:down))
    end

    acciones
end
```

```julia
acc8=[GWaction8(:up),GWaction8(:down),GWaction8(:right),GWaction8(:left),
    GWaction8(:upR),GWaction8(:upL),GWaction8(:downR),GWaction8(:downL)]

mdp=GW(ren,col,obj,ini,v,acc8,est)

episode_limit = 500
ϵ= 0.1
α=0.5
runs=1

figure_6_3(mdp, episode_limit,ϵ,α,runs)
```

## Ejercicio 6.10 

Stochastic Wind. Es el mismo ejercicio anterior con 8 posibles acciones, solo que ahora, con probabilidad de 1/3 el viento sera una unidad mas fuerte, con 1/3 de probabilidad sera una unidad menos fuerte y con 1/3 de probabilidad se mantendra igual.

```julia
struct GWact8Rand <:Accion
    a
    function GWact8Rand(a)
        if a!=:up && a!=:down && a!=:left && a!=:right && a!=:upR && a!=:upL && a!=:downR && a!=:downL
            error("acción inválida")
        else new(a)
        end
    end
end
```

```julia
function step(gw::GW,s::GWstate, a ::GWact8Rand)
    
    if gw.wind[s.y]>0
        r=rand()
        if r<=0.3333333333
            int=gw.wind[s.y]+1
        elseif 0.3333333333 < r <= 0.6666666666
            int=gw.wind[s.y]+1
        else
            int=gw.wind[s.y]
        end
    else int=0 end
    
    
    if a.a == :up 
        return GWstate(max(s.x - 1 - int, 1), s.y)
    elseif a.a == :down
        return GWstate(max(min(s.x + 1 - int, gw.rows), 1), s.y)
    elseif a.a == :left
        return GWstate(max(s.x - int, 1), max(s.y - 1, 1))
    elseif a.a == :right
        return GWstate(max(s.x - int, 1), min(s.y + 1, gw.columns))
        
    elseif a.a == :upR
        return GWstate(max(s.x  - 1- int, 1), min(s.y + 1, gw.columns))
    elseif a.a == :upL
        return GWstate(max(s.x  - 1- int, 1), max(s.y - 1, 1))
    elseif a.a == :downR
        return GWstate(max(min(s.x + 1 - int, gw.rows), 1), min(s.y + 1, gw.columns))
    elseif a.a == :downL
        return GWstate(max(min(s.x + 1 - int, gw.rows), 1), max(s.y - 1, 1))
    end
end
```

```julia
#acciones legales
function aLegales(mdp::GW,s::GWstate)

    acciones=[]

    if s.y > 1
        push!(acciones,GWact8Rand(:left))
        if s.x>1 push!(acciones,GWact8Rand(:upL)) end
        if s.x<mdp.rows push!(acciones,GWact8Rand(:downL)) end
    end
    if s.y<mdp.columns
        push!(acciones,GWact8Rand(:right))
        if s.x>1 push!(acciones,GWact8Rand(:upR)) end
        if s.x<mdp.rows push!(acciones,GWact8Rand(:downR)) end
    end
    if s.x >1
        push!(acciones,GWact8Rand(:up))
    end
    if s.x<mdp.rows
        push!(acciones,GWact8Rand(:down))
    end

    acciones
end
```

```julia
acc8=[GWact8Rand(:up),GWact8Rand(:down),GWact8Rand(:right),GWact8Rand(:left),
    GWact8Rand(:upR),GWact8Rand(:upL),GWact8Rand(:downR),GWact8Rand(:downL)]

mdp=GW(ren,col,obj,ini,v,acc8,est)

episode_limit = 500
ϵ= 0.1
α=0.5
runs=1

figure_6_3(mdp, episode_limit,ϵ,α,runs)
```
