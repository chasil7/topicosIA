```julia
using OpenAIGym
using Reinforce
```

```julia
global q_value=Dict()
global politica=Dict()
global α = 0.5
global ϵ = 0.1
```

```julia
env = GymEnv(:Acrobot, :v1)
```

```julia
struct Politica1<: AbstractPolicy end
function Reinforce.action(policy::Politica1, r, s, A)
    
    if s in keys(politica)
        """opciones=Dict()
        for (sx,a) in keys(q_value)
            if sx==s
               opciones[a]= q_value[(s,a)]
            end
        end
        action=findmax(opciones)[2]
        politica[s]=action"""
        return politica[s]
    else
        #println("no esta")
        a=rand(A)
        #politica[s]=a
        return a
    end
    
end
policy = Politica1()
```

```julia
struct greedy<: AbstractPolicy end
function Reinforce.action(policy::greedy, r, s, A,ϵ)
    
    if rand()<= ϵ
        action = rand(A)
    else
        # choose an action based on epsilon-greedy algorithm
        opciones=Dict()
        for (sx,a) in keys(q_value)
            if sx==s
               opciones[a]= q_value[(s,a)]
            end
        end
        action=findmax(opciones)[2]
    end
    action
end
policy_greedy= Politica1()
```

```julia
#Aqui deberia correr cada episodio y tener acceso a las acciones que se van tomando, los rewards etc
ep = Episode(env,policy_greedy)
```

```julia
#Intento de Sarsa
for i ∈ 1:20
    T=0
    R=0.0
    Reinforce.reset!(env)
    for (s, a, r, sp) in ep
        #if Reinforce.finished(env,sp)

        if (s,a) ∉ keys(q_value)
            q_value[(s,a)]=0.0
        end

        if !finished(env,sp)
            next_a = Reinforce.action(policy, r, sp, Reinforce.actions(env,sp))    
            if (sp,next_a) ∉ keys(q_value)
                q_value[(sp,next_a)]=0.0
            end
            q_value[(s,a)] += α * (r + q_value[(sp, next_a)] - q_value[(s,a)])

        else
            r2=0.0 #El reward de un estado terminal para el modelo de Acrobot es 0  
            q_value[(s,a)] += α * (r+r2 - q_value[(s,a)])
        end
        
        R+=r
        T+=1

        #@info("State $s, action $a, reward $r")
    end
    @info("Episode $i finished after $T steps. Total reward: $R")
end
```

```julia
struct RP<: AbstractPolicy end
function Reinforce.action(policy::RP, r, s, A)

    if Reinforce.finished(env) println("Termine") end
    reward=-2
    
    for a in collect(Reinforce.actions(env, s))
        r,s_ =step!(s,a)
        if r>reward 
            reward=r
            action=a
        end
    end
    action
end
```
